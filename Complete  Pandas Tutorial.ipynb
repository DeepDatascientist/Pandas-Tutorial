# Filtering dataframe on a condition | Masking using boolean series

```python
df.head()
```

```python
# To check out of data which match was played in hyderabad use the fuction below

# we also call it as masking where we create varialble called mask and store data in it but for now directly see it

df['city']=='Hyderabad'

# we call it as boolean series because against index it gives boolean values that is true and false
```

```python
# To only fetch the True value we use masking that is creating variable call mask and storing the condintion filtering
    
mask=df['city']=='Hyderabad'    


df[mask]

# Now it is only giving me data for the matches held hyderabad
```

```python
# To perform count how many matches played at Hyderabad we use function below 

df[mask].shape[0]

# 49 matches has been played at Hyderabad
```

```python
# Defining the function to give me city count i will create function

def get_city_counts(city):
    
    mask=df['city']==city
    return df[mask].shape[0]    
```

```python
# Now i can simply use function to get the count 

get_city_counts('Pune')

# 32 matches has been played in pune
```

```python
get_city_counts('Bangalore')

# 66 Matches has been played at Bangalore
```

```python
# Performing complex problem where i want to call out city and matches played after 2010

# here i will create 2 mask that is mask 1 and mask 2 and pass the condition

mask1 = df['city']=='Hyderabad'
mask2 = df['date']>'2010-01-01'

```

```python
df[mask1 & mask2] 

# here we are using bit & because in bolean we dont use normal and operator
```

```python
# Lets have the counts

df[mask1 & mask2].shape[0] 

# 42 Matches has been played at Hyderabad after 2010
```

```python
df.head()
```
